```python
import rasterio
import numpy as np
from rasterio import features
from shapely.geometry import Point
import geopandas as gpd
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import dijkstra

#1KM Buffer around roads
```

```python
cost_raster_path = os.getenv('cost')
with rasterio.open(cost_raster_path) as src:
    cost_array = src.read(1)  # Read the first band
    transform = src.transform
    crs = src.crs
    nodata = src.nodata
    height = src.height
    width = src.width

hospitals_gdf = gpd.read_file(os.getenv('surge-capacity'), layer='hospitalssurgecapacity__clipped').to_crs(crs)  # ensure same CRS as raster
```

```python
hospitals_gdf.plot()
```

```python
hospital_rcs = []
for idx, row in hospitals_gdf.iterrows():
    x, y = row.geometry.x, row.geometry.y
    r, c = ~transform * (x, y) 
    r_int, c_int = int(np.floor(r)), int(np.floor(c))
    if 0 <= r_int < height and 0 <= c_int < width:
        hospital_rcs.append((r_int, c_int))

hospital_rcs = list(set(hospital_rcs))  

aoi_gdf = gpd.read_file(os.getenv('aoishp')).to_crs(crs)

aoi_mask = features.geometry_mask(aoi_gdf.geometry, transform=transform, invert=True, out_shape=(height,width))

cost_array[~aoi_mask] = np.inf  
```

```python
def cell_index(r, c, width):
    return r * width + c

def neighbors(r, c, height, width):
    # 4-directional moves
    for nr, nc in [(r-1,c), (r+1,c), (r,c-1), (r,c+1)]:
        if 0 <= nr < height and 0 <= nc < width:
            yield nr, nc
```

```python
hospital_info = []

for idx, row in hospitals_gdf.iterrows():
    x_h, y_h = row.geometry.x, row.geometry.y
    # Convert map coords to row,col
    hr, hc = ~transform * (x_h, y_h)
    hr, hc = int(hr), int(hc)
    if 0 <= hr < height and 0 <= hc < width and np.isfinite(cost_array[hr, hc]):
        surge = row.get("norm_sc", 0)
        hospital_info.append((hr, hc, surge, x_h, y_h))  # store hospital coords and surge

# Build the graph from the cost array
row_indices = []
col_indices = []
weights = []

for r in range(height):
    for c in range(width):
        if np.isfinite(cost_array[r,c]):
            current_idx = cell_index(r, c, width)
            # Add edges to neighbors
            for nr, nc in neighbors(r,c,height,width):
                if np.isfinite(cost_array[nr,nc]):
                    # cost to move from cell to neighbor = (cost(current)+cost(neighbor))/2
                    w = (cost_array[r,c] + cost_array[nr,nc]) / 2.0
                    row_indices.append(current_idx)
                    col_indices.append(cell_index(nr,nc,width))
                    weights.append(w)
                    
graph = csr_matrix((weights, (row_indices, col_indices)), shape=(height*width, height*width))

# We'll compute distances from each hospital separately
final_dist = np.full(height*width, np.inf, dtype=np.float32)

# Precompute coordinates for each cell for the 2km buffer check
# This might be memory-heavy for large rasters; if so, consider a more efficient approach.
rows = np.arange(height)
cols = np.arange(width)
# Create 2D arrays of x,y coordinates
xx, yy = np.meshgrid(cols, rows)
# Apply the transform to get coordinates in map units
# x_map = x_origin + col*pixel_width
# y_map = y_origin + row*pixel_height
# For convenience, we can use rasterio.transform * to convert:
# But since we need all cells, let's do a vectorized approach:
x_map = transform.c + xx * transform.a + yy * transform.b
y_map = transform.f + xx * transform.d + yy * transform.e

# Distance threshold
distance_threshold = 2000.0  # 2 km

for (hr, hc, surge, x_h, y_h) in hospital_info:
    # Run Dijkstra from this hospital:
    # indices = cell_index(hr,hc,width) gives the hospital node
    temp_dist, predecessors = dijkstra(csgraph=graph, directed=False, indices=cell_index(hr,hc,width), 
                                       return_predecessors=True)
    
    # Adjust distances based on surge capacity
    # Higher surge capacity -> lower distances
    # Example formula: temp_dist = temp_dist / (1 + surge)
    temp_dist = temp_dist / (1 + surge)

    # Identify cells within 2 km of the hospital:
    # Compute squared distance to avoid sqrt where possible:
    dx = x_map - x_h
    dy = y_map - y_h
    dist_sq = dx*dx + dy*dy
    within_2km = dist_sq <= (distance_threshold * distance_threshold)
    
    # Further reduce cost for cells within 2 km buffer
    # For example, halve the distances within 2 km:
    temp_dist_2d = temp_dist.reshape((height, width))
    temp_dist_2d[within_2km] = temp_dist_2d[within_2km] * 0.5
    
    # Flatten again for merging with final_dist
    temp_dist = temp_dist_2d.flatten()

    # Combine results with final_dist
    final_dist = np.minimum(final_dist, temp_dist)

# Reshape final result back to 2D
cost_distance = final_dist.reshape((height, width))

# Replace infinite distances (unreachable areas) with a nodata value
nodata_value = -9999
cost_distance[~np.isfinite(cost_distance)] = nodata_value

# Write the output raster
out_meta = {
    'driver': 'GTiff',
    'height': height,
    'width': width,
    'count': 1,
    'dtype': 'float32',
    'crs': crs,
    'transform': transform,
    'nodata': nodata_value
}

output_path = os.getenv('accessability')
with rasterio.open(output_path, 'w', **out_meta) as dst:
    dst.write(cost_distance, 1)
```

```python


row_indices = []
col_indices = []
weights = []

for r in range(height):
    for c in range(width):
        if np.isfinite(cost_array[r,c]):
            current_idx = cell_index(r, c, width)
            # Add edges to neighbors
            for nr, nc in neighbors(r,c,height,width):
                if np.isfinite(cost_array[nr,nc]):
                    # cost to move from cell to neighbor = (cost(current) + cost(neighbor))/2
                    w = (cost_array[r,c] + cost_array[nr,nc]) / 2.0
                    row_indices.append(current_idx)
                    col_indices.append(cell_index(nr,nc,width))
                    weights.append(w)
                    
graph = csr_matrix((weights, (row_indices, col_indices)), shape=(height*width, height*width))

# 4. Multi-source shortest path
# We need to specify the starting points for Dijkstra.
# Dijkstra in scipy.sparse.csgraph does not directly accept multiple sources, 
# but we can handle it by initializing distances and running a custom loop or by
# setting the initial distance for all hospital cells to 0 and using a trick:

dist = np.full(height*width, np.inf, dtype=np.float32)

# Set hospital cells distance to 0
for (hr, hc) in hospital_rcs:
    dist[cell_index(hr,hc,width)] = 0

# Now run Dijkstra with these distances as initial conditions
# We can run once per hospital or build a "virtual" super-source:
# The simplest approach is to run Dijkstra once per hospital and take the minimum,
# but that might be expensive for many hospitals.

# For efficiency, let's do a simple approach: If there are few hospital sources:
final_dist = np.full_like(dist, np.inf)
for (hr,hc) in hospital_rcs:
    temp_dist, predecessors = dijkstra(csgraph=graph, directed=False, indices=cell_index(hr,hc,width), return_predecessors=True)
    final_dist = np.minimum(final_dist, temp_dist)

# final_dist now contains the accumulated cost to the nearest hospital cell for each cell

# 5. Reshape the result back to raster shape
cost_distance = final_dist.reshape((height, width))

# Replace infinite distances (unreachable areas) with a nodata value
cost_distance[~np.isfinite(cost_distance)] = -9999  # or another chosen NoData value

# 6. Write the output raster
out_meta = {
    'driver': 'GTiff',
    'height': height,
    'width': width,
    'count': 1,
    'dtype': 'float32',
    'crs': crs,
    'transform': transform,
    'nodata': -9999
}

output_path = "accessibility_cost_distance.tif"
with rasterio.open(output_path, 'w', **out_meta) as dst:
    dst.write(cost_distance, 1)
```
